#### Bar Chart

```python
a1 = [2,4,6,8,10]
a2 = [1,3,5,7,9]
b0 = [3,6,9,12,15]
b1 = [2,5,8,11,14]
```

#### Histogram

```python
c=np.random.randint(1,300,size=(1050))
bins=np.arange(10,300,10)
```

```python
plt.hist(c,bins,histtype='bar',rwidth=0.7)
plt.show()
```

```python
plt.hist(c,bins,histtype='bar',rwidth=0.6,label='Frequency Distribution',color='k')
plt.legend()
plt.grid(True,color='r')
plt.title("Histogram")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.show()
```

#### Scatter plot

```python
d=np.random.randint(1,300,size=500)
e=np.random.randint(1,50,size=500)
plt.scatter(d,e)
plt.show()
```

```python
plt.scatter(d,e,color='k',s=20,marker='x')
plt.grid(True,color='r')
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

#### Pie Plot

```python
Slice=[10,7,3,2,2]
Activities=['Work','Sleep','Movie & Youtube','Study','Other']
```

```python
plt.pie(Slice,labels=Activities,startangle=180,shadow=True,autopct='%1.1f%%',radius=1)
plt.title("Pie Chart")
plt.show()
```

```python
plt.pie(Slice,labels=Activities,startangle=180,shadow=True,autopct='%1.1f%%',radius=1,explode=[0,0.1,0,0,0])
plt.title("Pie Chart")
plt.show()
```

#### Stackplot

```python
g =  [1, 2, 3, 4, 5, 6, 7]
h1 = [1, 1, 2, 3, 5, 8, 13]
h2 = [0, 2, 4, 6, 8, 10, 12]
h3 = [1, 3, 5, 7, 9, 11, 13]
h = np.vstack([h1, h2, h3])
labels = ["Fibonacci ", "Evens", "Odds"]
```

fig, ax = plt.subplots()
ax.stackplot(g, h1,h2, labels=labels)
ax.legend(loc='upper left')
plt.show()

```python
fig, ax = plt.subplots()
ax.stackplot(g, h,labels=labels)
ax.legend(loc='upper left')
plt.show()
```

#### Subplot

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
#plt.margins(x=-0.25, y=-0.25)

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
#plt.margins(x=-0.25, y=-0.25)

plt.subplot(2, 2, 3)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.subplot(2, 2, 2)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
#plt.margins(x=-0.25, y=-0.25)

plt.subplot(2, 2, 4)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```
