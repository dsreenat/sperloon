```python
#creating a pandas serires
pds = pd.Series([1,2,3,4,5,6])
print(type(pds))
pds
```

```python
#retriving the index of the serires,
print("printing value of the series pds_1: \n")
print(pds_1,"\n")

#iloc used to select an element based on its location
print("2nd Index Location using iloc ",pds_1.iloc[2],"\n")

#alternate option
print("2nd Index Location ",pds_1[2],"\n")

#shape of the series
print("Shape of the series ",pds_1.shape,"\n")

#size of the series
print("Size of the series ",pds_1.size,"\n")

print("printing first 3 elements of pds_1")
pds_1[0:3]
```

```python
#Creating a series with a custom index
a=["a","b","c","d","e","f","g"]
b=[1,2,3,4,5,6,7]
#a=[a,b,c,d,e,f,g]

my_var = pd.Series(a,index=b)
print("series with index of numbers starting from 1 :")
print(my_var,"\n")

#note :default index is from 0, here we made a custom index from 1
#lets see another example.

my_var2 = pd.Series(b, index =a)
print("series with index of letters starting from a : \n")
print(my_var2,"\n")
```

```python
# how to access an element in a series with the index value. eg as follows
print("2nd element of series my_var :",my_var[2],"\n")

print("element at loc 'g' of series my_var2 :",my_var2["g"])
```

#### Join

```python
Join_1= pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})

Join_2 = pd.DataFrame({'key': ['K0', 'K1', 'K2','D1'],
'B': ['B0', 'B1', 'B2','D1']})

print(Join_1,"\n",Join_2)
```

```python
#joining two tables. by default left join
Join_1.join(Join_2, lsuffix='_J1', rsuffix='_J2')
```

```python
#Inner Join
Join_1.join(Join_2, lsuffix='_J1', rsuffix='_J2',how='inner')
```

```python
#outer
Join_1.join(Join_2, lsuffix='_J1', rsuffix='_J2',how='outer')
```

```python
#Right
Join_1.join(Join_2, lsuffix='_J1', rsuffix='_J2',how='right')
```

how{‘left’, ‘right’, ‘outer’, ‘inner’}, default ‘left’
How to handle the operation of the two objects.

left: use calling frame’s index (or column if on is specified)

right: use other’s index.

outer: form union of calling frame’s index (or column if on is specified) with other’s index, and sort it. lexicographically.

inner: form intersection of calling frame’s index (or column if on is specified) with other’s index, preserving the order of the calling’s one.
